```python
#enter direct string that has form of a list
def stringToList(string):
    return ast.literal_eval(string)


#splits list of dictionary with id and name as key into 2d list with id values in first list and name values in second list
def listOfDictToColumns(list):
        genres_id = []
        genre_names = []
        for i in list:
            for k, v in i.items():
                if k == "id":
                    genres_id.append(v)
                elif k == "name":
                    genre_names.append(v)
        return [genres_id, genre_names]
```

```python
#error happens due to rerunning this cell, can only be run once
clean_sub_df.genres = clean_sub_df.genres.map(lambda x: stringToList(x))
```

```python
clean_sub_df.genres = clean_sub_df.genres.map(lambda x: listOfDictToColumns(x))
```
