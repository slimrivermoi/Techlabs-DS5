# Parse ['Cast'] and ['Crew']

```python
import pandas as pd
import ast

credits_df = pd.read_csv("../../source_data/credits.csv")
credits_df.head()
```

### Parse ['Cast']

```python
# Function to parse the string and extract names
def extract_names(cast_string):
    # Safely evaluate the string to a Python object
    cast_list = ast.literal_eval(cast_string)
    # Extract names
    names = [cast['name'] for cast in cast_list]
    return names

# Apply the function to the DataFrame
credits_df['cast_names'] = credits_df['cast'].apply(extract_names)

# Display the DataFrame
credits_df.head(20)
```

### Parse ['crew']

There are many roles in crews such as director, writer, executive producer, sound, stage, etc. 
We agreed to use only director as it is the most exposed role compared to other titles. 

In the code below, we will parse the director's name and gender. 

```python
# to see the details of crew data within a role
credits_df['crew'][0]

## Note: add 'job': 'Director' as condition
# extract 'name': 'Joe Johnston', 'gender': 2
```

```python
# Function to parse the string and extract names
def extract_names(crew_string):
    # Safely evaluate the string to a Python object
    crew_list = ast.literal_eval(crew_string)
    # Extract names to a dictionary with condition if job = director
    director = [{crew['name']:crew['gender'] for crew in crew_list if crew['job']=='Director'}]
    return director

# Apply the function to the DataFrame
credits_df['director'] = credits_df['crew'].apply(extract_names)

# Display the DataFrame
credits_df.head(20)
```
