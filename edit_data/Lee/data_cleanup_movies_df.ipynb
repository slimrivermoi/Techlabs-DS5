### Step 2: Data Preparation

Null Values & Duplications
- do missing values exist, and how often?

```python
# Null values table
movies_df.isna().sum()
```

Remove duplicates in Id, Original Title, Vote_Counts

```python
# to capture duplicated rows based on ['Id']
movies_df.loc[movies_df.duplicated(subset=['id'])]
```

```python
#inspect the duplicated rows
# notes: inspected a few and they are all identical except in column "Popularity"
movies_df.query('id == "4912"')
```

```python
# add ~ to exclude the duplicated rows in our df
# Exclude if it is identical in 3 columns (id, Vote_Count, Original_Title)
movies_df = movies_df.loc[~movies_df.duplicated(subset=['id', 'vote_count', 'original_title'])].reset_index(drop=True).copy()

#now there are less rows than the original set (28 rows less)
print(movies_df.shape)

#double check if the ID is still duplicated
movies_df.query('id == "4912"')
```

```python
#inspect if there are further movies with duplicate ID 
movies_df.loc[movies_df.duplicated(subset=['id'])]

#result shows 2 movie (id:15028 Clockstoppers, 10991 Pokémon 3: The Movie)
```

```python
#Query these movies to remove the entry based on higher popularity score
movies_df.query('id == "15028"')

#remove movie with index 33722
filt = ~((movies_df['id'] == "15028") & (movies_df.index == 33722))
movies_df = movies_df.loc[filt]
print(movies_df.shape)
```

```python
#Query these movies to remove the entry based on higher popularity score
movies_df.query('id == "10991"')

#remove movie with index 33722
filt = ~((movies_df['id'] == "10991") & (movies_df.index == 44795))
movies_df = movies_df.loc[filt]
print(movies_df.shape)
```

Remove Adult Movies

```python
# Filter to include only NON-ADULT movies
# This will also remove 3 entries that are misaligned columns. 
filt = ( movies_df['adult'] == 'False' )
movies_df = movies_df.loc[filt].copy()
movies_df.shape
```

Remove Non-Released Movies

```python
# filter
filt = (movies_df['status'] == 'Released')
movies_df = movies_df.loc[filt].copy()
movies_df.shape
```

Include only non-video Movies

```python
filt2 = (movies_df['video'] == False)
movies_df = movies_df.loc[filt2]
print(movies_df.shape)
```

Remove rows that have inaccurate data

```python
#no records when search by vote_average is null
movies_df['vote_average'].isna().sum()
```

Include only columns we need.
Reorder the columns.

```python
movies_df.columns
```

```python
#copy a new df
movies_df = movies_df[['id','original_title', 'original_language','spoken_languages',
       'budget', 'revenue', 'runtime', 'release_date',
        'production_companies','production_countries', 
       'imdb_id', 'popularity', 'vote_average', 'vote_count','belongs_to_collection', 
       'tagline','overview', 'genres']].copy()
```

### Update Dtypes 

```python
movies_df.dtypes
```

```python

#['budget'] & ['popularity'] -> float
# ['release_date'] -> datetime
#NOTE: there are some invalid string data in ['budget'] and ['popularity']. Use method error='coerce' to set invalid parsing as NaN.

import datetime

movies_df['release_date'] = pd.to_datetime(movies_df['release_date'], errors='coerce')
movies_df['budget'] = pd.to_numeric(movies_df['budget'], errors = 'coerce')
movies_df['popularity'] = pd.to_numeric(movies_df['popularity'], errors = 'coerce')
```

```python
# convert ['budget'] from integer to float
movies_df['budget'] = movies_df['budget'].astype(float)
```

```python
#confirm the dtypes are now correctly updated
movies_df.dtypes
```

```python
# inspect the data again to make sure we did not lose the values in the series (eg. date)
movies_df.head()
```

### Rename columns

```python
movies_df.columns
```

```python
movies_df = movies_df.rename(columns={'id': 'Id',
                        'original_title': 'Original_Title',
                        'original_language': 'Original_Language',
                        'spoken_languages': 'Spoken_Languages',
                        'budget': 'Budget',
                        'revenue': 'Revenue',
                        'runtime': 'Runtime',
                        'release_date': 'Release_Date',
                        'production_companies': 'Production_Companies',
                        'production_countries': 'Production_Countries',
                        'Imdb_id': 'Imdb_Id',
                        'popularity': 'Popularity',
                        'vote_average': 'Vote_Average',
                        'vote_count': 'Vote_Count', 
                        'belongs_to_collection': 'Belongs_To_Collection',
                        'tagline': 'Tagline', 
                        'overview': 'Overview',
                        'genres': 'Genres'})
```
