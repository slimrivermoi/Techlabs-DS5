```python
import pandas as pd

movies_df = pd.read_csv("../../edit_data/cleaned_data/movies.csv")
movies_df.head()
```

#### Parse ['Genres']

```python
import json

# First - convert ' to "
genres = movies_df.Genres.replace("\'","\"")

# Loop the list in the cell to convert to json data. 
for i in range(len(genres)):
    genres[i] = movies_df.Genres[i].replace("\'","\"")
    genres[i] = json.loads(genres[i])


# parse genres JSON data and save it back as list
all_genres = []

for each_film in genres:
    new_list=[]
    for item in each_film:
        new_item = item['name']
        new_list.append(new_item)
    each_film = new_list
    # print(each_film)
    all_genres.append(each_film)

#assign the list as ['Genres']
movies_df['Genres'] = all_genres
    
```

#### Parse ['Spoken_Languages']

```python
# First - convert ' to "
languages = movies_df.Spoken_Languages.replace("\'","\"")

# Loop the list in the cell to convert to json data. 
 
for i in range(len(languages)):
    languages[i] = movies_df.Spoken_Languages[i].replace("\'","\"")
    try:
        languages[i] = json.loads(languages[i])
    except (json.JSONDecodeError, TypeError, KeyError) as e:
            pass
```

```python
# parse genres JSON data and save it back as list
all_languages = []

for each_film in languages:
    new_list=[]
    for item in each_film:
        try:
            new_item = item['name']
            new_list.append(new_item)
        except (json.JSONDecodeError, TypeError, KeyError) as e:
            pass
    each_film = new_list
    
    all_languages.append(each_film)

movies_df['Spoken_Languages'] = all_languages
```
