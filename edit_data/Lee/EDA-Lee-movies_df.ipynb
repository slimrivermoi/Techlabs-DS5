Remove duplicates in Id, Original Title, Vote_Counts

```python
# to capture duplicated rows based on ['Id']
movies_df.loc[movies_df.duplicated(subset=['id'])]
```

```python
#inspect the duplicated rows
# notes: inspected a few and they are all identical except in column "Popularity"
movies_df.query('id == "4912"')
```

```python
# add ~ to exclude the duplicated rows in our df
# Exclude if it is identical in 3 columns (id, Vote_Count, Original_Title)
movies_df = movies_df.loc[~movies_df.duplicated(subset=['id', 'vote_count', 'original_title'])].reset_index(drop=True).copy()

#now there are less rows than the original set (28 rows less)
print(movies_df.shape)

#double check if the ID is still duplicated
movies_df.query('id == "4912"')
```

- Copy DF with only columns need
- update dtypes 

```python
#copy a new df
movies_df = movies_df[['id','original_title','budget', 'revenue', 
       'release_date',  'runtime', 'vote_average', 'vote_count',
       'imdb_id', 'popularity','tagline', 'overview', 'production_companies',
       'production_countries','genres', 'spoken_languages']].copy()
```

```python
#update dtype for columns
#budget, popularity - float
#release_date - datetime
#NOTE: there are some invalid string data in ['budget'] and ['popularity']. Use method error='coerce' to set invalid parsing as NaN.

import datetime

movies_df['release_date'] = pd.to_datetime(movies_df['release_date'], errors='coerce')
movies_df['budget'] = pd.to_numeric(movies_df['budget'], errors = 'coerce')
movies_df['popularity'] = pd.to_numeric(movies_df['popularity'], errors = 'coerce')
```

Parsing the Genres Column

```python
movies_df['Genres'].head()
```

```python
#Parse the Genres column
# First - convert ' to "
# Look the list in the cell to convert to json data. 

import json
genres = movies_df.Genres.replace("\'","\"")

for i in range(len(genres)):
    genres[i] = movies_df.Genres[i].replace("\'","\"")
    # print(genres[i])
    genres[i] = json.loads(genres[i])
```

```python
genres
```

```python
# test the data in the nested list
print(genres[0][0]['name'])

# create a list to capture all unique value of genres
genres_list = []

#create a set to capture unique IDs
unique_ids = set()

# check how many genres are there
for each_film in genres:
    for item in each_film:
        if item['id'] not in unique_ids:
            entry = {'id': item['id'],
                     'name': item['name'],
                     }
            
            genres_list.append(entry)
            unique_ids.add(item['id'])

```
