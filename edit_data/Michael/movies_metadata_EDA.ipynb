```python
#isn't reusable/dynamic, has to be changed
def listOfDictToColumns(list):
        genres_id = []
        genre_names = []
        for i in list:
            #genres_id_sublist = []
            #genre_names_sublist = []
            for k, v in i.items():
                if k == "id":
                    genres_id.append(v)
                elif k == "name":
                    genre_names.append(v)
            #genres_id.append(genres_id_sublist)
            #genre_names.append(genres_name_sublist)
        return [genres_id, genre_names]
```

```python
listOfDictToColumns(df.genres[0])
```

```python
df.genres = df.genres.map(lambda x: listOfDictToColumns(x))
```
