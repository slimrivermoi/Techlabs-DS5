```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
import numpy as np
```

```python
df = pd.read_csv("../cleaned_data/movies_main.csv")
```

```python
df
```

```python
df["Overview"] = df["Overview"].astype("string")
```

```python
df["Overview"].dtype
```

```python
overview_series = df["Overview"].loc[(df["Overview"] != "NaN")]
overview_series.count()
```

```python
# in general the formula looks something like this

# TF - Term frequency = (number of repetitions of words) / (number of words in sentence) ----- frequency of word in one text 
# IDF - Inverse Document Frequency = log( (number of texts) / (number of texts containing the word) ) ------ inverse frequency of word in all texts, the closer the value to zero is the more often the word appears in all texts
# weighted value = TF * IDF

#in basic terms the more often the words occur the lower the rating, filler words get a low rating this way and can be filtered 
#also contains zero values for every used word in all texts but not occurring in specific text
```

```python
#initialize the vectorizer object
vectorizer = TfidfVectorizer(stop_words = "english") #possible keys stop_words = "english"
#this is where there will be created a vector for every value of column["overview"] which contains the "Term Frequency Inverse Document Frequency" for every word

vectors = vectorizer.fit_transform(overview_series.to_list())
#gets example vector for a movie and presents values
first_vector = vectors[0]
dfTEST = pd.DataFrame(first_vector.T.todense(), index= vectorizer.get_feature_names_out(), columns=["tfidf"]).sort_values(by=["tfidf"],ascending=False)
```

```python
df_short = df.copy()[["Id", "Overview"]]
df_short
```

```python
#Code works but is extremely slow, would take about 30 mins to compute all rows

liste = []
for i in range(df_short.head(10).shape[0]):
    keywords_rating = pd.DataFrame(vectors[i].T.todense(), index= vectorizer.get_feature_names_out(), columns=["tfidf"]).sort_values(by=["tfidf"],ascending=False)
    #df_short["keywords"] = list(dfTEST.loc[dfTEST["tfidf"]>0.15].index)
    liste.append(list(keywords_rating.loc[dfTEST["tfidf"]>0.15].index))
```

```python
# Get feature names and convert vectors to sparse matrix 
feature_names = vectorizer.get_feature_names_out()

liste = []
# Loop through each row (document vector)
for i in range(vectors.shape[0]):
    # Get the non-zero indices and their corresponding TF-IDF scores
    row = vectors[i]
    nonzero_indices = row.indices
    nonzero_data = row.data

    # Filter for TF-IDF scores greater than 0.15
    high_tfidf_indices = nonzero_indices[nonzero_data > 0.15]
    keywords = feature_names[high_tfidf_indices]
    
    # Append the list of keywords to the 'liste' list
    liste.append(keywords.tolist())
liste
```

```python
series_keywords = pd.Series(liste)
```

```python
df_short["keywords_tfidf"] = series_keywords
```
